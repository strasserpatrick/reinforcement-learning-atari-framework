```python
from pathlib import Path

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from helper import get_multiple_runs_dict, add_plot_line
```

```python
double_dqn_df = get_multiple_runs_dict(Path("data/Freeway/double_dqn"))
regular_dqn_df = get_multiple_runs_dict(Path("data/Freeway/regular_dqn"))
mini_rainbow_df = get_multiple_runs_dict(Path("data/Freeway/mini_rainbow"))
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

add_plot_line(double_dqn_df, ax, "Double DQN", "blue")
add_plot_line(regular_dqn_df, ax, "Regular DQN", "orange")
add_plot_line(mini_rainbow_df, ax, "Mini Rainbow", "green")

# add title
ax.set_title("Freeway")

# add x and y labels
ax.set_xlabel("Steps")
ax.set_ylabel("Average Validation Reward")

# add xtickslabel
ax.set_xticklabels([f"{int(tick / 1000)}k" for tick in ax.get_xticks()])

ax.legend(loc='upper right')
plt.show()
```
